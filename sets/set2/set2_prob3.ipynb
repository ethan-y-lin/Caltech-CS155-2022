```python
import numpy as np
from matplotlib import pyplot as plt
from numpy import genfromtxt
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
%matplotlib inline
```

## Load data

```python
# Upload problem3data.txt to a 'data' folder in Google Colab, using the "Files"
# tab on the left.
# (You can also use urllib.request.urlretrieve, as in set 1)
train_file = 'data/problem3data.txt'
train_data = genfromtxt(train_file, delimiter='\t')

y_train = train_data[:, 9]
X_train = train_data[:, :9]
```

```python
# Problem A
# i. 
a_array = np.linspace(0.01, 3, 30)
w_arr = []
for a in a_array:
    lasso = Lasso(a)
    lasso.fit(X_train, y_train)
    w_arr.append(lasso.coef_)

w_arr = np.array(w_arr)
plt.figure()
count = 1
for w_i in w_arr.T:
    plt.plot(a_array, w_i, label=("w_" + str(count)))
    count+=1

plt.xlabel("a")
plt.ylabel("Weights")
plt.legend()

```

```python
#ii
a_array = np.arange(1, 10001)
w_arr = []
for a in a_array:
    ridge = Ridge(a)
    ridge.fit(X_train, y_train)
    w_arr.append(ridge.coef_)

w_arr = np.array(w_arr)
plt.figure()
count = 1
for w_i in w_arr.T:
    plt.plot(a_array, w_i, label=("w_" + str(count)))
    count+=1

plt.xlabel("a")
plt.ylabel("Weights")
plt.legend()
```
